### Column description

The features selected for this database come from the accelerometer and gyroscope 3-axial raw signals tAcc-XYZ and tGyro-XYZ. These time domain signals (prefix 't' to denote time) were captured at a constant rate of 50 Hz. Then they were filtered using a median filter and a 3rd order low pass Butterworth filter with a corner frequency of 20 Hz to remove noise. Similarly, the acceleration signal was then separated into body and gravity acceleration signals (tBodyAcc-XYZ and tGravityAcc-XYZ) using another low pass Butterworth filter with a corner frequency of 0.3 Hz. 


Subsequently, the body linear acceleration and angular velocity were derived in time to obtain Jerk signals (tBodyAccJerk-XYZ and tBodyGyroJerk-XYZ). Also the magnitude of these three-dimensional signals were calculated using the Euclidean norm (tBodyAccMag, tGravityAccMag, tBodyAccJerkMag, tBodyGyroMag, tBodyGyroJerkMag). 


Finally a Fast Fourier Transform (FFT) was applied to some of these signals producing fBodyAcc-XYZ, fBodyAccJerk-XYZ, fBodyGyro-XYZ, fBodyAccJerkMag, fBodyGyroMag, fBodyGyroJerkMag. (Note the 'f' to indicate frequency domain signals). 


These signals were used to estimate variables of the feature vector for each pattern:  
'-XYZ' is used to denote 3-axial signals in the X, Y and Z directions.

```python
def uniquify(df_columns):
    seen = set()

    for item in df_columns:
        fudge = 1
        newitem = item

        while newitem in seen:
            fudge += 1
            newitem = "{}_{}".format(item, fudge)

        yield newitem
        seen.add(newitem)
```

t-distributed stochastic neighbor embedding (**TSNE**) is a statistical method for visualizing high-dimensional data by giving each datapoint a location in a two or three-dimensional map. ## Outliery

We defined function which we will use later to transform outliers in each column. We defined outliers as values that are lower than q1 (first quantile) minus 1.5 times IQR (interquartile range) or higher than q3 + 1.5 * IQR. Having detected oultiers in each column we transformed them to outlier boundary.

```python
from sklearn.base import BaseEstimator, TransformerMixin

class OutlierRemover(BaseEstimator,TransformerMixin):
    def __init__(self,factor=1.5):
        self.factor = factor
        
    def outlier_detector(self,X,y=None):
        X = pd.Series(X).copy()
        q1 = X.quantile(0.25)
        q3 = X.quantile(0.75)
        iqr = q3 - q1
        self.lower_bound.append(q1 - (self.factor * iqr))
        self.upper_bound.append(q3 + (self.factor * iqr))

    def fit(self,X,y=None):
        self.lower_bound = []
        self.upper_bound = []
        X.apply(self.outlier_detector)
        return self
    
    def transform(self,X,y=None):
        X = pd.DataFrame(X).copy()
        for i in range(X.shape[1]):
            x = X.iloc[:, i].copy()
            x[(x < self.lower_bound[i])] = self.lower_bound[i]
            x[(x > self.upper_bound[i])] = self.upper_bound[i]
            X.iloc[:, i] = x
        return X
    
outlier_remover = OutlierRemover()
```

## Deleting columns with higher correlation than 0.7 and lower predictivity.

In our dataset there are grups of features that derive from one measured physical quantity and represent different statistic metrics. Therefore some of them are highly correlated and can be dumped. To choose which of them to get rid of best we decided to first calculate feature importances and then out of correlated columns delete these with lesser importance.`

```python
X_train = train.drop('activity', axis=1)
y_train = train['activity']
```

```python
outlier_remover = OutlierRemover()

outlier_remover.fit(X_train)

X_train = outlier_remover.transform(X_train)
```

```python
import xgboost as xgb
from xgboost import XGBClassifier
```

```python
model = XGBClassifier(verbosity=0)
model.fit(X_train, y_train)
```

```python
importances = pd.DataFrame(data={
    'name': X_train.columns,
    'importance': model.feature_importances_
})
importances = importances.sort_values(by='importance', ascending=False)

plt.figure(figsize=(12,8))
plt.bar(x=importances['name'][:10], height=importances['importance'][:10], color='#087E8B')
plt.title('Top 10 important features', size=20)
plt.xticks(rotation='vertical')
plt.show()
```

```python
corr_matrix = X_train.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
```

```python
def col_to_drop(dataset, upper):
    to_drop = []
    for column in upper.columns:
        if any(upper[column] > .7):
            correlated = upper[upper[column] > .7].index
            for col in correlated:
                if dataset[dataset['name'] == col]['importance'].values[0] < dataset[dataset['name'] == column]['importance'].values[0]:
                    to_drop.append(col)
                else:
                    to_drop.append(column)
    return to_drop
```

```python
to_drop = col_to_drop(importances, upper)
```

```python
train.drop(to_drop, axis=1, inplace=True)
```

```python
train.reset_index(drop=True, inplace = True)
```

```python
X, y = train.drop(['activity'], axis=1), train['activity']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import StackingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
def get_models():
    models = dict()
    models['logisreg'] = LogisticRegression()
    models['randforclas'] = RandomForestClassifier()
    models['svc'] = SVC(probability=True)
    models['xgboost'] = XGBClassifier(verbosity = 0)
    models['knn'] = KNeighborsClassifier()
    models['gaussNB'] = GaussianNB()
    return models
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_validate
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
scoring = {'acc': 'accuracy',
           'f1': 'f1_macro',
           'rec': 'recall_macro',
           'prec' : 'precision_macro',
           'roc' : 'roc_auc_ovr'}

def evaluate_model(model, X, y):
    
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=36851234)
    clf = make_pipeline(OutlierRemover(), StandardScaler(), model)
    scores = cross_validate(clf, X, y, scoring=scoring, cv=cv, n_jobs=-1, error_score='raise')

    return scores
```

```python
models = get_models()
colnames = ['fit_time', 'score_time', 'test_acc', 'test_f1', 'test_rec', 'test_prec', 'test_roc', 'model']
df = pd.DataFrame(columns = colnames)

results, names = list(), list()

for name, model in models.items():
    scores = evaluate_model(model, X, y)
    dt = pd.DataFrame(scores)
    dt['model'] = name
    df = pd.concat([df, dt], axis=0)
```

```python
df2 = df.groupby('model').mean()
df2.drop(['fit_time','score_time'], axis=1, inplace=True)
df2.sort_values('test_acc', ascending=False).style.highlight_max(color = 'lightgreen', axis = 0)
```

As we can see the best performing model out of these 6 models is  XGBClassifier having highest score in all 5 evaluation metrics.

Let's plot confusion matrices for all our tested models.

```python
X_train = train.drop('activity', axis=1)
y_train = train['activity']

pipeline = make_pipeline(OutlierRemover(), StandardScaler())
pipeline.fit(X_train)
X_train = pipeline.transform(X_train)

X_test = test.drop('activity', axis=1)
X_test = pipeline.transform(X_test)
y_test = test['activity']
```

```python
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix

classifiers = [LogisticRegression(), RandomForestClassifier(), 
               SVC(probability=True), XGBClassifier(verbosity = 0), 
               KNeighborsClassifier(), GaussianNB()]

for cls in classifiers:
    cls.fit(X_train, y_train)
    
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(16,16))

for cls, ax in zip(classifiers, axes.flatten()):
    plot_confusion_matrix(cls, 
                          X_test, 
                          y_test, 
                          ax=ax, 
                          cmap='Blues',
                         display_labels=y_test.unique(),
                         xticks_rotation='vertical')
    ax.title.set_text(type(cls).__name__)
plt.tight_layout()  
plt.subplots_adjust(bottom=0.15,
                   hspace=1,
                   wspace = 0)
plt.show()
```

## Further evaluation

```python
model = XGBClassifier(verbosity=0)
model.fit(X_train, y_train)

y_proba = model.predict_proba(X_test)
y_pred = model.predict(X_test)
```

```python
print("Accuracy is {0:.6f}".format(accuracy_score(y_test, y_pred)))
print("Precision is {0:.6f}".format(precision_score(y_test, y_pred, average='macro')))
print("Recall is {0:.6f}".format(recall_score(y_test, y_pred, average='macro')))
print("F1_score is {0:.6f}".format(f1_score(y_test, y_pred, average='macro')))
```

Now we will try to create a stacking classifier out of all models that we tested earlier (except for GaussianNB, because it didn't perform as well as others).

### Stacking

```python
estimators = [
    ('logisreg',LogisticRegression()),
    ('randforclas',RandomForestClassifier()),
    ('svc',SVC(probability=True)),
    ('xgboost',XGBClassifier(verbosity = 0)),
    ('knn', KNeighborsClassifier())
]
model = StackingClassifier(
     estimators=estimators, final_estimator=LogisticRegression()
)
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print("Accuracy is {0:.6f}".format(accuracy_score(y_test, y_pred)))
print("Precision is {0:.6f}".format(precision_score(y_test, y_pred, average='macro')))
print("Recall is {0:.6f}".format(recall_score(y_test, y_pred, average='macro')))
print("F1_score is {0:.6f}".format(f1_score(y_test, y_pred, average='macro')))
```

```python
cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
fig, ax = plt.subplots(figsize=(10,10))
disp.plot(ax=ax, cmap='Blues')
plt.xticks(rotation=45)
plt.show()
```
